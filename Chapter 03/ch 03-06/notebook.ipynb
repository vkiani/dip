```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.ones((512,512,3), np.uint8)
img = img * 255

img = cv.rectangle(img, (200,100), (500,150), (0,0,255), 3)

img = cv.circle(img, (400,250), 50, (255,0,0), -1)

img = cv.line(img, (10,10), (500,500), (0,0,255), 3)

img = cv.putText(img, 'Image Processing', (20,400), cv.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 2, cv.LINE_AA)
img = cv.line(img, (10,420), (330,420), (0,0,0), 3)

img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```
