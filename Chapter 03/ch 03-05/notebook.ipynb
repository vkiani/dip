```python
import cv2 as cv
import matplotlib.pyplot as plt

img_gray = cv.imread('house.tif', cv.IMREAD_GRAYSCALE)

img_crop = img_gray[10:300, 10:400]

cv.imwrite('sample.png', img_crop)

plt.figure()

plt.subplot(1,2,1)
plt.imshow(img_gray, cmap="gray")

plt.subplot(1,2,2)
plt.imshow(img_crop, cmap="gray")

plt.show()
```
