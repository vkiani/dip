```python
import cv2 as cv
import matplotlib.pyplot as plt

img_gray = cv.imread('house.tif', cv.IMREAD_GRAYSCALE)
img_color = cv.imread('peppers_color.tif', cv.IMREAD_COLOR)

img_rgb = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)

plt.figure()

plt.subplot(1,2,1)
plt.imshow(img_gray, cmap="gray")

plt.subplot(1,2,2)
plt.imshow(img_rgb)

plt.show()
```
