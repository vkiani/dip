```python
import cv2 as cv
import matplotlib.pyplot as plt

left = cv.imread('tsukuba_left.tiff', cv.IMREAD_GRAYSCALE)
right = cv.imread('tsukuba_right.tiff', cv.IMREAD_GRAYSCALE)

stereo = cv.StereoBM_create(numDisparities=16, blockSize=15)

disparity = stereo.compute(left, right)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1), plt.imshow(left, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(right, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.figure(figsize=(10,10))

plt.subplot(1,2,1), plt.imshow(left, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(disparity, cmap='gray')
plt.xticks([]), plt.yticks([])
```
