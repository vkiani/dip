```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('peppers_color.tif', cv.IMREAD_COLOR)

rows, cols, channels = img.shape

M = cv.getRotationMatrix2D((cols/2, rows/2), 30, 1)

res = cv.warpAffine(img, M, img.shape[0:2])

plt.subplot(1,2,1), plt.imshow(img[:,:,::-1]), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(res[:,:,::-1]), plt.xticks([]), plt.yticks([])
```
