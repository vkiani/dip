```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('peppers_color.tif', cv.IMREAD_COLOR)

res_1 = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)
res_2 = cv.rotate(img, cv.ROTATE_180)

plt.subplot(1,3,1), plt.imshow(img[:,:,::-1]), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2), plt.imshow(res_1[:,:,::-1]), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3), plt.imshow(res_2[:,:,::-1]), plt.xticks([]), plt.yticks([])
```
