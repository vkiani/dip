```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('book.jpg', cv.IMREAD_COLOR)

plt.figure;
plt.imshow(img[:,:,::-1])

rows, cols = (400, 300)

pts1 = np.float32([[848,6],[1306,159],[897,647],[366,387]])
pts2 = np.float32([[0,0],[cols,0],[cols,rows],[0,rows]])

M = cv.getPerspectiveTransform(pts1,pts2)

res = cv.warpPerspective(img, M, [cols,rows])

plt.figure()
plt.subplot(1,2,1), plt.imshow(img[:,:,::-1]), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(res[:,:,::-1]), plt.xticks([]), plt.yticks([])
```
