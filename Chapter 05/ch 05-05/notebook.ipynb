```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('noisy_rect.jpg', cv.IMREAD_GRAYSCALE)

plt.figure()
plt.hist(img.ravel(), 50)
plt.show()

ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)

ret,thresh2 = cv.threshold(img,0,255, cv.THRESH_BINARY + cv.THRESH_OTSU)

print(ret)

plt.figure()
plt.subplot(1,3,1), plt.imshow(img,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2), plt.imshow(thresh1,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3), plt.imshow(thresh2,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```
