```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('peppers_color.tif', cv.IMREAD_COLOR)

res_1 = cv.resize(img, None, fx=2, fy=2, interpolation = cv.INTER_AREA)
res_2 = cv.resize(img, None, fx=2, fy=2, interpolation = cv.INTER_CUBIC)

plt.subplot(2,2,1), plt.imshow(res_1[:,:,::-1])
plt.subplot(2,2,2), plt.imshow(res_2[:,:,::-1])
plt.subplot(2,2,3), plt.imshow(res_1[0:100,0:100,::-1])
plt.subplot(2,2,4), plt.imshow(res_2[0:100,0:100,::-1])
```
