```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('great_books.jpg', cv.IMREAD_COLOR)

plt.figure;
plt.imshow(img)

rows, cols = (400, 300)

pts1 = np.float32([[579,213],[819,369],[576,743]])
pts2 = np.float32([[0,0],[cols,0],[cols,rows]])

M = cv.getAffineTransform(pts1,pts2)

res = cv.warpAffine(img, M, [cols,rows])

plt.figure()
plt.subplot(1,2,1), plt.imshow(img[:,:,::-1]), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(res[:,:,::-1]), plt.xticks([]), plt.yticks([])
```
