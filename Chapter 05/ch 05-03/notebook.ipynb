```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('cameraman.tif', cv.IMREAD_GRAYSCALE)

ret,thresh1 = cv.threshold(img,80,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,80,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,80,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,80,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,80,255,cv.THRESH_TOZERO_INV)

titles = ['Original image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
