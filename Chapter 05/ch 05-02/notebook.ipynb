```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random

img = cv.imread('peppers_color.tif', cv.IMREAD_COLOR)

res = np.zeros(img.shape, img.dtype)

p = 0.30    # probablity of noise

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        
        r = random.random()
        
        if r < p/2:
            res[i,j] = [0,0,0]
        elif r < p:
            res[i,j] = [255,255,255]
        else:
            res[i,j] = img[i,j]

plt.subplot(1,2,1), plt.imshow(img[:,:,::-1])
plt.subplot(1,2,2), plt.imshow(res[:,:,::-1])
```
