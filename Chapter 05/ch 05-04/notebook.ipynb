```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('sudoku.png', cv.IMREAD_GRAYSCALE)

ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)

thresh2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
               cv.THRESH_BINARY,11,2)

thresh3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
               cv.THRESH_BINARY,11,2)


plt.subplot(2,2,1), plt.imshow(img,cmap='gray')
plt.subplot(2,2,2), plt.imshow(thresh1,cmap='gray')
plt.subplot(2,2,3), plt.imshow(thresh2,cmap='gray')
plt.subplot(2,2,4), plt.imshow(thresh3,cmap='gray')
```
