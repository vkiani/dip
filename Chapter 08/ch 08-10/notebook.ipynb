```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('tennis.jpg', cv.IMREAD_COLOR)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

ret, mask = cv.threshold(gray, 200, 255, cv.THRESH_BINARY)

minLineLength = 100
maxLineGap = 10

lines = cv.HoughLinesP(mask, 1, np.pi/180, 100, minLineLength, maxLineGap)

res = img.copy()

for line in lines[0:10]:
    for x1,y1,x2,y2 in line:
        cv.line(res, (x1,y1), (x2,y2), (0,0,255), 2)

plt.figure(figsize=(15,40)) 

plt.subplot(1,2,1), plt.imshow(img[:,:,::-1])
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(res[:,:,::-1])
plt.xticks([]), plt.yticks([])
```
