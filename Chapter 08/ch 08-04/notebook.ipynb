```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('tsukuba.jpg', cv.IMREAD_GRAYSCALE)

clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

res = clahe.apply(img)

plt.subplot(1,2,1), plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(res, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])
```
