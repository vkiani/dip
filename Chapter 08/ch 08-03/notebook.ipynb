```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Hawkes.jpg', cv.IMREAD_GRAYSCALE)

res = cv.equalizeHist(img)

hist_img = cv.calcHist([img],[0],None,[50],[0,256])
hist_res = cv.calcHist([res],[0],None,[50],[0,256])

plt.subplot(2,2,1), plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2), plt.imshow(res, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3), plt.plot(hist_img)

plt.subplot(2,2,4), plt.plot(hist_res)
```
