```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('ronaldo.jpg', cv.IMREAD_COLOR)

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

roi_mask = np.zeros((img.shape[0],img.shape[1]), dtype='uint8')
roi_mask[300:400,100:300] = 255

roi_hist = cv.calcHist([hsv], [0, 1], roi_mask, [180, 256], [0, 180, 0, 256])

res = cv.calcBackProject([hsv], [0,1], roi_hist, [0,180,0,256], 1)

plt.figure(figsize=(15,40)) 

plt.subplot(1,2,1), plt.imshow(img[:,:,::-1])
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(res > 0, cmap='gray')
plt.xticks([]), plt.yticks([])
```
