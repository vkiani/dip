```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('mario.jpg', cv.IMREAD_COLOR)

coin = img[81:97,72:83,:]

res = cv.matchTemplate(img,coin,cv.TM_CCOEFF)

print('min score: ', np.min(res))
print('max score: ', np.max(res))

plt.figure(figsize=(15,40)) 

plt.subplot(1,3,1), plt.imshow(img[:,:,::-1])
plt.xticks([]), plt.yticks([])

plt.subplot(1,3,2), plt.imshow(res, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(1,3,3), plt.imshow(res > 1.5e6, cmap='gray')
plt.xticks([]), plt.yticks([])
```
