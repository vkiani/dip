```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('peppers.tif', cv.IMREAD_GRAYSCALE)

hist_full = cv.calcHist([img],[0],None,[50],[0,256])

mask = np.zeros((img.shape[0],img.shape[1]), np.uint8)
mask[0:150, 100:400] = 255
masked_img = cv.bitwise_and(img,img,mask = mask)

hist_mask = cv.calcHist([masked_img],[0],mask,[50],[0,256])

plt.subplot(2,2,1), plt.imshow(img, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2), plt.plot(hist_full)

plt.subplot(2,2,3), plt.imshow(masked_img, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,4), plt.plot(hist_mask)
```
