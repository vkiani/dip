```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('coins.jpg', cv.IMREAD_COLOR)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

circles = cv.HoughCircles(gray, cv.HOUGH_GRADIENT, 1, 20,
                          param1=100,param2=50,minRadius=40,maxRadius=80)

res = img.copy()

circles = np.uint16(np.around(circles))

for circle in circles[0,0:10]:
    cx, cy, radius = circle
    cv.circle(res, (cx,cy), radius, (0,255,0), 2)
    cv.circle(res, (cx,cy), 2, (0,0,255), 3)

plt.figure(figsize=(15,40)) 

plt.subplot(1,2,1), plt.imshow(img[:,:,::-1])
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(res[:,:,::-1])
plt.xticks([]), plt.yticks([])
```
