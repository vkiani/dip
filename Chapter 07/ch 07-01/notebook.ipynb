```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import math

img = cv.imread('j_noisy.png', cv.IMREAD_GRAYSCALE)

th, img = cv.threshold(img, 128, 255, cv.THRESH_BINARY)

kernel = np.ones((5,5), np.uint8)

res = cv.erode(img, kernel, iterations = 1)

plt.subplot(1,2,1), plt.imshow(img, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(res, cmap='gray')
plt.xticks([]), plt.yticks([])
```
