```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('rice.png', cv.IMREAD_GRAYSCALE)

kernel = np.ones((5,5), np.uint8)

res1 = cv.erode(img, kernel)
res2 = cv.dilate(img, kernel)
res3 = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

plt.subplot(2,2,1), plt.imshow(img, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2), plt.imshow(res1, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3), plt.imshow(res2, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,4), plt.imshow(res3, cmap='gray')
plt.xticks([]), plt.yticks([])
```
