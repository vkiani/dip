```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('peppers_color.tif', cv.IMREAD_COLOR)

# access to color value of one pixel
print( img[100,200] )

# change color value of one pixel
img[100,200] = [255,0,0]

print( img[100,200] )

img_rgb_1 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# change color value of a rectangular region of image
img[100:110,200:210] = [255,0,0]

img_rgb_2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)


plt.figure()

plt.subplot(1,2,1)
plt.imshow(img_rgb_1)

plt.subplot(1,2,2)
plt.imshow(img_rgb_2)

plt.show()
```
