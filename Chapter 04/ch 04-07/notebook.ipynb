```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('cameraman.tif', cv.IMREAD_GRAYSCALE)
logo = cv.imread('logo.png', cv.IMREAD_GRAYSCALE)

rows, cols = logo.shape

compose_1 = img.copy()

part_1 = compose_1[10:10+rows, 20:20+cols]

part_1[:] = cv.addWeighted(part_1, 0.7, logo, 0.3, 0)

plt.subplot(1,2,1), plt.imshow(img, cmap='gray')
plt.subplot(1,2,2), plt.imshow(compose_1, cmap='gray')
```
