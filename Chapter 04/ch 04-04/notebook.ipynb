```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('peppers_color.tif', cv.IMREAD_COLOR)

print(img.shape)

blue, green, red = cv.split(img)

mask = green > 100

img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.figure()

plt.subplot(1,2,1)
plt.imshow(img_rgb)

plt.subplot(1,2,2)
plt.imshow(mask, cmap='gray')

plt.show()
```
