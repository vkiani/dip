```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('cameraman.tif', cv.IMREAD_GRAYSCALE)
logo = cv.imread('logo.png', cv.IMREAD_GRAYSCALE)

rows, cols = logo.shape

compose_1 = img.copy()
compose_2 = img.copy()

part_1 = compose_1[10:10+rows, 20:20+cols]
part_2 = compose_2[10:10+rows, 20:20+cols]

part_1[:] = part_1 + logo
part_2[:] = cv.add(part_2, logo)

plt.subplot(1,2,1), plt.imshow(compose_1, cmap='gray')
plt.subplot(1,2,2), plt.imshow(compose_2, cmap='gray')
```
