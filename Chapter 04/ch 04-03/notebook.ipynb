```python
import cv2 as cv
import matplotlib.pyplot as plt

img_orig = cv.imread('ronaldo.jpg', cv.IMREAD_COLOR)

img_copy = img_orig.copy()

ball = img_copy[365:425, 405:465]

img_copy[365:425, 200:260] = ball

img_orig_rgb = cv.cvtColor(img_orig, cv.COLOR_BGR2RGB)
img_copy_rgb = cv.cvtColor(img_copy, cv.COLOR_BGR2RGB)

plt.figure()

plt.subplot(1,2,1)
plt.imshow(img_orig_rgb)
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2)
plt.imshow(img_copy_rgb)
plt.xticks([]), plt.yticks([])

plt.show()
```
