```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('cameraman.tif', cv.IMREAD_GRAYSCALE)

dx = cv.Sobel(img,-1,1,0,ksize=3)
dy = cv.Sobel(img,-1,0,1,ksize=3)

edge_sobel = cv.add(dx,dy)
edge_sobel = edge_sobel > 50

dx = cv.Scharr(img,-1,1,0)
dy = cv.Scharr(img,-1,0,1)

edge_scharr = cv.add(dx,dy)
edge_scharr = edge_scharr > 150

edge_laplacian = cv.Laplacian(img,-1)
edge_laplacian = edge_laplacian > 6

plt.subplot(2,2,1), plt.imshow(img, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2), plt.imshow(edge_sobel, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3), plt.imshow(edge_scharr, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4), plt.imshow(edge_laplacian, cmap='gray'), plt.xticks([]), plt.yticks([])
```
