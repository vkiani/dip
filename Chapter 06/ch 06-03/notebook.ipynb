```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('house.tif', cv.IMREAD_GRAYSCALE)

probability = np.random.rand(img.shape[0], img.shape[1])

noise = img.copy()
noise[probability < 0.05] = 0
noise[probability > 0.95] = 255

res = cv.GaussianBlur(noise,(11,11),0)

plt.subplot(1,2,1), plt.imshow(noise, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(res, cmap='gray'), plt.xticks([]), plt.yticks([])
```
