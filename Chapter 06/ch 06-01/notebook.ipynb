```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('house.tif', cv.IMREAD_GRAYSCALE)

kernel = np.ones((10,10),np.float32)/100

res = cv.filter2D(img,-1,kernel)

plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(res, cmap='gray'), plt.xticks([]), plt.yticks([])
```
