```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import math

img = cv.imread('cameraman_dark.tif', cv.IMREAD_GRAYSCALE)

img_float = np.float32(img)

dft = cv.dft(img_float, flags=cv.DFT_COMPLEX_OUTPUT)

mag, phase = cv.cartToPolar(dft[:,:,0], dft[:,:,1])

mag = cv.pow(mag, 1.2)

real, imag = cv.polarToCart(mag, phase)

res_dft = cv.merge([real, imag])

res_complex = cv.idft(res_dft)

res = cv.magnitude(res_complex[:,:,0], res_complex[:,:,1])

# plot results
plt.subplot(1,2,1), plt.imshow(img, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(res, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.show() 
```
