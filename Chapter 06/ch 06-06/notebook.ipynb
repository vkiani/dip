```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('cameraman.tif', cv.IMREAD_GRAYSCALE)

dx = cv.Sobel(img,-1,1,0,ksize=3)
dy = cv.Sobel(img,-1,0,1,ksize=3)

edge = cv.add(dx,dy)

plt.subplot(2,2,1), plt.imshow(img, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2), plt.imshow(dx, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3), plt.imshow(dy, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4), plt.imshow(edge, cmap='gray'), plt.xticks([]), plt.yticks([])
```
